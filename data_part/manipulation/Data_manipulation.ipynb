```python
def get_titles(name):
    first_name = name.split(",")[-1]
    title= first_name.split(".")[0]
    return title.strip(" ")
```

```python
get_titles("oli,MR . sam")
```

```python

titanic_df.Name.apply(get_titles)
```

```python
titanic_df['title']=titanic_df.Name.apply(get_titles)
```
